# 01 - Exploratory Data Analysis (EDA)

Objetivo: entender la distribución del target, la calidad de los datos y relaciones clave.

```python
# Importar librerías y cargar datos
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv("../data/train.csv")
df.head()
```

```python
# Revisar columnas
df.columns
```

```python
# Resumen de tipos y nulos
df.info()
```

```python
# Estadísticas descriptivas
df.describe()
```

```python
# Distribución de SalePrice
sns.histplot(df["SalePrice"], kde=True)
plt.title("Distribución de precios")
plt.show()
```

```python
# Correlación con SalePrice
corr = df.corr(numeric_only=True)
corr["SalePrice"].sort_values(ascending=False)
```

```python
# Relación GrLivArea vs SalePrice
sns.scatterplot(x="GrLivArea", y="SalePrice", data=df)
plt.show()
```

```python
# Transformar SalePrice y ver distribución
df["SalePriceLog"] = np.log1p(df["SalePrice"])
sns.histplot(df["SalePriceLog"], kde=True)
plt.title("Distribución logarítmica")
plt.show()
```

```python
# Detectar outliers en SalePrice
sns.boxplot(x=df["SalePrice"])
```

```python
# Revisar valores nulos
nulls = df.isnull().sum().sort_values(ascending=False)
nulls = nulls[nulls > 0]
print(nulls)
```
